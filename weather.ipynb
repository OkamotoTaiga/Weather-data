```python
fig = plt.figure(figsize=(12,8))
sns.distplot(data['maxtemp-mean'])
sns.distplot(data['mintemp-mean'])
plt.grid()
```

```python
fig = plt.figure(figsize=(15, 12))
plt.plot(data['year'],data['maxtemp-mean'],c='r',label='max temp')
plt.plot(data['year'],data['temp-mean'],c='g',label='mean temp')
plt.plot(data['year'],data['mintemp-mean'],c='b',label='min temp')
plt.legend(loc='best')
plt.show()
```

```python
data['maxtemp-ave'] = data['maxtemp-mean'].rolling(window=5,min_periods=1,center=True).mean()
data['temp-ave'] = data['temp-mean'].rolling(window=5,min_periods=1,center=True).mean()
data['mintemp-ave'] = data['mintemp-mean'].rolling(window=5,min_periods=1,center=True).mean()
```

```python
data = data.dropna()
data.head()
```

```python
fig = plt.figure(figsize=(15, 12))
plt.scatter(data['year'],data['maxtemp-ave'],c='r')
plt.scatter(data['year'],data['mintemp-ave'],c='b')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 12))
sns.regplot(x='year',y='maxtemp-ave',data=data)
sns.regplot(x='year',y='mintemp-ave',data=data)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#最高気温
X = data.loc[:,['year']].as_matrix()
Y = data['maxtemp-ave'].as_matrix()
clf = LinearRegression()
clf.fit(X,Y)

# 回帰係数と切片の抽出
a1 = clf.coef_
b1 = clf.intercept_  
# 回帰係数
print('最高気温')
print("回帰係数:", a1)
print("切片:", b1) 
print("決定係数:", clf.score(X, Y))
print('\n')

#平均気温
Y = data['temp-ave'].as_matrix()
clf = LinearRegression()
clf.fit(X,Y)

# 回帰係数と切片の抽出
a2 = clf.coef_
b2 = clf.intercept_  
# 回帰係数
print('平均気温')
print("回帰係数:", a2)
print("切片:", b2) 
print("決定係数:", clf.score(X, Y))
print('\n')

#最低気温
Y = data['mintemp-ave'].as_matrix()
clf = LinearRegression()
clf.fit(X,Y)

# 回帰係数と切片の抽出
a3 = clf.coef_
b3 = clf.intercept_  
# 回帰係数
print('最低気温')
print("回帰係数:", a3)
print("切片:", b3) 
print("決定係数:", clf.score(X, Y))

#線形回帰直線の定義
def maxtemp(x):
    return a1*x + b1

def meantemp(x):
    return a2*x + b2

def mintemp(x):
    return a3*x + b3
```

```python
fig = plt.figure(figsize=(30, 12))
plt.plot([i for i in range(1870,2500)],[maxtemp(i) for i in range(1870,2500)],c='black',lw=3)
plt.plot([i for i in range(1870,2500)],[mintemp(i) for i in range(1870,2500)],c='black',lw=3)
plt.scatter(data['year'],data['maxtemp-ave'],c='r',label='maxtemp')
plt.scatter(data['year'],data['mintemp-ave'],c='b',label='min temp')
plt.legend(loc='best')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 12))
plt.plot(X,maxtemp(X),c='black',lw=3)
plt.plot(X,mintemp(X),c='black',lw=3)
plt.scatter(data['year'],data['maxtemp-ave'],c='r',label='maxtemp')
plt.scatter(data['year'],data['mintemp-ave'],c='b',label='min temp')
plt.legend(loc='best')
plt.show()
```
